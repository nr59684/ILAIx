```python
# -----------------------------
# Define a Prediction Function for SHAP
# -----------------------------
def predict_fn(texts):
    """
    Converts a list of texts into model predictions.
    Returns probabilities computed via sigmoid.
    """
    encoded = tokenizer(
        texts,
        return_tensors="pt",
        padding="max_length",
        truncation=True,
        max_length=512
    )
    # Move tensors to the proper device
    encoded = {k: v.to(device) for k, v in encoded.items()}
    outputs = model(**encoded)
    logits = outputs.logits
    probs = torch.sigmoid(logits)
    return probs.detach().cpu().numpy()
```

```python
# --- Preprocessing Functions ---
def remove_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text(separator=" ")

def clean_special_chars(text):
    text = re.sub(r"http\S+", "", text)
    cleaned_text = re.sub(r"[^a-zA-Z0-9\s.,!?;:'\"-]", "", text)
    return cleaned_text

def normalize_whitespace(text):
    cleaned_text = " ".join(text.split())
    return cleaned_text.strip()

def preprocess_text(text):
    text = remove_html_tags(text)
    text = clean_special_chars(text)
    text = normalize_whitespace(text)
    text = text.lower()
    return text
```

```python
model_directory = "../../model/Bert3.0"

preprocessed_license_text = preprocess_text(license_text)

model = AutoModelForSequenceClassification.from_pretrained(model_directory)
tokenizer = AutoTokenizer.from_pretrained(model_directory)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

loaded_data = np.load("../mlb.npz", allow_pickle=True)
loaded_classes = loaded_data["classes"]
mlb = MultiLabelBinarizer()
mlb.classes_ = loaded_classes
num_labels = len(mlb.classes_)
```

```python
def predictor(input_ids, attention_mask=None):
    """
    Prediction function for DeepExplainer.
    Handles BERT's inputs and outputs.
    """
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
        probabilities = torch.sigmoid(outputs.logits) #we will do attribution for each logit.
    return probabilities

def get_top_sentences_with_deepexplainer(text, preprocessed_text, predicted_labels, shap_values, top_k=3):
    """
    Gets top-k sentences based on aggregated DeepExplainer SHAP values.
    """
    sentences = sent_tokenize(text)
    preprocessed_sentences = [preprocess_text(s) for s in sentences]
    result_dict = {}

    for label_index, label in enumerate(predicted_labels[0]):
        sentence_scores = []
        # Aggregate SHAP values *per token*
        token_shap_values = shap_values[label_index]

        for sent_idx, sent in enumerate(preprocessed_sentences):
            tokens = tokenizer.tokenize(sent)
            token_ids = tokenizer.convert_tokens_to_ids(tokens)
            input_ids = tokenizer.encode(preprocessed_text, add_special_tokens=True, truncation=True, max_length=512)
            indices = [input_ids.index(token_id) for token_id in token_ids if token_id in input_ids]

            if indices:
                sentence_score = np.sum(token_shap_values[indices])
            else:
                sentence_score = 0.0
            sentence_scores.append((sentences[sent_idx], sentence_score))

        sentence_scores.sort(key=lambda x: x[1], reverse=True)
        top_sentences = [sentence for sentence, _ in sentence_scores[:top_k]]
        result_dict[label] = top_sentences

    return result_dict
```
